# DATA ANALYSIS OF CREDIT CARD ACCEPTANCE

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset

```python
csv_file= 'credit_card.csv'
data= pd.read_csv(csv_file)

data.head()
```

```python
data.info()
```

# Data description

```python
data.describe()
```

## Finding null value

```python
plt.figure(figsize=(10, 5))
plt.title('Heatmap of null value')
sns.heatmap(data.isnull(), cbar=False)
```

# Correlation of data features

```python
# corrilation

data.corr()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Correlation graph')
sns.heatmap(data.corr(), cbar= True, annot= True, annot_kws={'size':15})
plt.show()
```

## Missing value Treatment

# How many people have credit card and not? 

```python
plt.figure(figsize=(10,5))
plt.title('Number of card accepted and not')

sns.countplot(data.card)
plt.xlabel('card')
plt.ylabel('No of people')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Majorcards VS no. of people having card')
c= data[data.card=='yes']
sns.countplot(c.majorcards)
plt.xlabel('major cards')
plt.ylabel('No of people')
plt.show()
```

# How many card holder and without card people have their reports

```python
plt.figure(figsize=(10, 5))
plt.title('Card VS Reports')
plt.bar(data.card, data.reports)
plt.xlabel('card')
plt.ylabel('No of reports')
plt.show()
```

# How many people are the owner

```python
# Owner VS no. of people

plt.figure(figsize=(10, 5))
plt.title('Owner VS no. of people')

sns.countplot(data.owner)
plt.xlabel('Owner')
plt.ylabel('No of people')

plt.show()
```

# How many of the owner with card and without card are selfemployed

```python
o= data[data.owner=='yes']
os= o[o.selfemp=='yes']
plt.title('Number of owner selfemployed VS card')

sns.countplot(os.card, data=data)
plt.xlabel('card')
plt.ylabel('Number of people')

plt.show()
```

# income difference between selfemployed owner and not owner selfemployed

```python
o= data[data.selfemp=='yes']
plt.title('income VS owner')

#sns.boxplot(x=o.owner,y=o.income, data=data,color='orange')
plt.bar(o.owner, o.income)
plt.xlabel('owner')
plt.ylabel('income')
```

# How many people are selfemployed and not

```python
# Selfemp VS no. of people

plt.figure(figsize=(10, 5))
plt.title('Selfemp VS no. of people')
sns.countplot(data.selfemp)

plt.xlabel('selfemployed')
plt.ylabel('No of people')

plt.show()
```

# How many people have their Reports having their card and majorcards

```python
c= data[data.card=='yes']
cm= c[c.majorcards==1]

plt.figure(figsize=(10,5))
plt.title('Reports VS peoples having Card and majorcards')
sns.countplot(cm.reports)
plt.xlabel('No of reports')
plt.ylabel('No of people')
plt.show()
```

# What is the active accounts status of the people who have card and major cards also

```python
c= data[data.card=='yes']
cm= c[c.majorcards==1]

plt.figure(figsize=(10,5))
plt.title('No of active accounts VS no. of people having Card and majorcards')
sns.countplot(cm.active)
plt.xlabel('No of active account')
plt.ylabel('No of people')
plt.show()
```
